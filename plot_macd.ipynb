```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output.csv')
```

```python
df.shape
```

```python
# Plotting the line chart for weekly
plt.figure(figsize=(18, 10))  # Adjust the figure size if needed
plt.plot(df['Date'], df['weekly_macd'], label='Weekly MACD')  # 'o' for markers, '-' for line style
plt.plot(df['Date'], df['weekly_signal'], label='Weekly Signal')  # 'o' for markers, '-' for line style
plt.title('Line Chart')  # Set the title of the plot
plt.xlabel('Date')  # Set the label for the x-axis
# plt.ylabel('Y-axis label')  # Set the label for the y-axis
plt.grid(axis='y')  # Show gridlines
plt.axhline(0, color='black', linewidth=2, label='Y=0') 
plt.legend()

# Add vertical lines based on 'switch_code'
for index, row in df.iterrows():
    if row['switch_code'] == 1:
        plt.axvline(row['Date'], color='red', linestyle='--', linewidth=2, label='Switch to 0')
    elif row['switch_code'] == 2:
        plt.axvline(row['Date'], color='green', linestyle='--', linewidth=2, label='Switch to 3')


plt.savefig('line_chart_macd_and_signal.png')
```

```python
# Plotting the line chart for monthly
plt.figure(figsize=(18, 10))  # Adjust the figure size if needed
plt.plot(df['Date'], df['monthly_macd'], label='Monthly MACD')  # 'o' for markers, '-' for line style
plt.plot(df['Date'], df['monthly_signal'], label='Monthly Signal')  # 'o' for markers, '-' for line style
plt.title('Line Chart')  # Set the title of the plot
plt.xlabel('Date')  # Set the label for the x-axis
# plt.ylabel('Y-axis label')  # Set the label for the y-axis
plt.grid(axis='y')  # Show gridlines
plt.axhline(0, color='black', linewidth=2, label='Y=0') 
plt.legend()

plt.savefig('line_chart_macd_and_signal.png')
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df['Date'] = pd.to_datetime(df['Date'])

# Chart 1: Weekly MACD and Signal
plt.figure(figsize=(18, 10))  # Adjust the figure size if needed
plt.subplot(2, 1, 1)  # 2 rows, 1 column, select the first subplot
plt.plot(df['Date'], df['weekly_macd'], label='Weekly MACD')  # 'o' for markers, '-' for line style
plt.plot(df['Date'], df['weekly_signal'], label='Weekly Signal')  # 'o' for markers, '-' for line style
plt.title('Weekly MACD and Signal')  # Set the title of the plot
plt.xlabel('Date')  # Set the label for the x-axis
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Set major ticks to months
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format x-axis to display month and year
plt.xticks(rotation=45) 

plt.grid(axis='y')  # Show gridlines
plt.axhline(0, color='black', linewidth=2, label='Y=0') 
plt.legend()
# Add vertical lines based on 'switch_code'
for index, row in df.iterrows():
    if row['switch_code'] == 1:
        plt.axvline(row['Date'], color='pink', linestyle='--', linewidth=2, label='only_weekly_neg')
    if row['switch_code'] == 2:
        plt.axvline(row['Date'], color='red', linestyle='--', linewidth=2, label='both_neg')
    if row['switch_code'] == 3:
        plt.axvline(row['Date'], color='lightgreen', linestyle='--', linewidth=2, label='only_weekly_pos')
    elif row['switch_code'] == 4:
        plt.axvline(row['Date'], color='green', linestyle='--', linewidth=2, label='both_pos')

# Chart 2: Monthly MACD and Signal
plt.subplot(2, 1, 2)  # 2 rows, 1 column, select the second subplot
plt.plot(df['Date'], df['monthly_macd'], label='Monthly MACD')  # 'o' for markers, '-' for line style
plt.plot(df['Date'], df['monthly_signal'], label='Monthly Signal')  # 'o' for markers, '-' for line style
plt.title('Monthly MACD and Signal')  # Set the title of the plot
plt.xlabel('Date')  # Set the label for the x-axis
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Set major ticks to months
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format x-axis to display month and year
plt.xticks(rotation=45) 
plt.grid(axis='y')  # Show gridlines
plt.axhline(0, color='black', linewidth=2, label='Y=0') 
plt.legend()

# Adjust layout to prevent clipping of titles
plt.tight_layout()



# Add vertical lines based on 'switch_code'
for index, row in df.iterrows():
    if row['switch_code'] == 1:
        plt.axvline(row['Date'], color='pink', linestyle='--', linewidth=2, label='only_weekly_neg')
    if row['switch_code'] == 2:
        plt.axvline(row['Date'], color='red', linestyle='--', linewidth=2, label='both_neg')
    if row['switch_code'] == 3:
        plt.axvline(row['Date'], color='lightgreen', linestyle='--', linewidth=2, label='only_weekly_pos')
    elif row['switch_code'] == 4:
        plt.axvline(row['Date'], color='green', linestyle='--', linewidth=2, label='both_pos')

# Show the plot
plt.show()
# Save the plot
plt.savefig('weekly_and_monthly_macd_and_signal.png')
```
