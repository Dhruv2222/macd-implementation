```python
# Using sub plots

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming df is your DataFrame and ticker_symbol is defined
df['Date'] = pd.to_datetime(df['Date'])

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(18, 15))

# Chart 1: Closing Price
axs[0].plot(df['Date'], df['Close'], label='Weekly MACD')  
axs[0].set_title('Closing Price - ' + ticker_symbol, color='gray')  
axs[0].set_xlabel('Date')  
axs[0].xaxis.set_major_locator(mdates.MonthLocator())  
axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  
axs[0].tick_params(rotation=45) 
axs[0].grid(axis='y')  
axs[0].axhline(0, color='black', linewidth=2, label='Y=0') 
axs[0].legend()

# Add vertical lines based on 'switch_code'
for index, row in df.iterrows():
    if row['switch_code'] == 1:
        axs[0].axvline(row['Date'], color='pink', linestyle='--', linewidth=2, label='only_weekly_neg')
    if row['switch_code'] == 2:
        axs[0].axvline(row['Date'], color='red', linestyle='--', linewidth=2, label='both_neg')
    if row['switch_code'] == 3:
        axs[0].axvline(row['Date'], color='lightgreen', linestyle='--', linewidth=2, label='only_weekly_pos')
    elif row['switch_code'] == 4:
        axs[0].axvline(row['Date'], color='green', linestyle='--', linewidth=2, label='both_pos')

# Chart 2: Weekly MACD and Signal
axs[1].plot(df['Date'], df['weekly_macd'], label='Weekly MACD')  
axs[1].plot(df['Date'], df['weekly_signal'], label='Weekly Signal')  
axs[1].set_title('Weekly MACD and Signal - ' + ticker_symbol, color='gray')  
axs[1].set_xlabel('Date')  
axs[1].xaxis.set_major_locator(mdates.MonthLocator())  
axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  
axs[1].tick_params(rotation=45) 
axs[1].grid(axis='y')  
axs[1].axhline(0, color='black', linewidth=2, label='Y=0') 
axs[1].legend()

# Add vertical lines based on 'switch_code'
for index, row in df.iterrows():
    if row['switch_code'] == 1:
        axs[1].axvline(row['Date'], color='pink', linestyle='--', linewidth=2, label='only_weekly_neg')
    if row['switch_code'] == 2:
        axs[1].axvline(row['Date'], color='red', linestyle='--', linewidth=2, label='both_neg')
    if row['switch_code'] == 3:
        axs[1].axvline(row['Date'], color='lightgreen', linestyle='--', linewidth=2, label='only_weekly_pos')
    elif row['switch_code'] == 4:
        axs[1].axvline(row['Date'], color='green', linestyle='--', linewidth=2, label='both_pos')

# Chart 3: ATR
axs[2].plot(df['Date'], df['atr'], label='Weekly MACD')  
axs[2].set_title('ATR - ' + ticker_symbol, color='gray')  
axs[2].set_xlabel('Date')  
axs[2].xaxis.set_major_locator(mdates.MonthLocator())  
axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  
axs[2].tick_params(rotation=45) 
axs[2].grid(axis='y')  
axs[2].legend()

# Add vertical lines based on 'switch_code'
for index, row in df.iterrows():
    if row['switch_code'] == 1:
        axs[2].axvline(row['Date'], color='pink', linestyle='--', linewidth=2, label='only_weekly_neg')
    if row['switch_code'] == 2:
        axs[2].axvline(row['Date'], color='red', linestyle='--', linewidth=2, label='both_neg')
    if row['switch_code'] == 3:
        axs[2].axvline(row['Date'], color='lightgreen', linestyle='--', linewidth=2, label='only_weekly_pos')
    elif row['switch_code'] == 4:
        axs[2].axvline(row['Date'], color='green', linestyle='--', linewidth=2, label='both_pos')

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure
plt.savefig('MSFT_3_plots_weekly.png')
```
