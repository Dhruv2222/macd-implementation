```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output.csv')
```

```python
df.shape
```

```python
# Plotting the line chart for daily
plt.figure(figsize=(18, 10))  # Adjust the figure size if needed
plt.plot(df['Date'], df['daily_macd'], label='Daily MACD')  # 'o' for markers, '-' for line style
plt.plot(df['Date'], df['daily_signal'], label='Daily Signal')  # 'o' for markers, '-' for line style
plt.title('Line Chart')  # Set the title of the plot
plt.xlabel('Date')  # Set the label for the x-axis
# plt.ylabel('Y-axis label')  # Set the label for the y-axis
plt.grid(axis='y')  # Show gridlines
plt.axhline(0, color='black', linewidth=2, label='Y=0') 
plt.legend()

plt.savefig('line_chart_macd_and_signal.png')
```

```python
# Plotting the line chart for weekly
plt.figure(figsize=(18, 10))  # Adjust the figure size if needed
plt.plot(df['Date'], df['weekly_macd'], label='Weekly MACD')  # 'o' for markers, '-' for line style
plt.plot(df['Date'], df['weekly_signal'], label='Weekly Signal')  # 'o' for markers, '-' for line style
plt.title('Line Chart')  # Set the title of the plot
plt.xlabel('Date')  # Set the label for the x-axis
# plt.ylabel('Y-axis label')  # Set the label for the y-axis
plt.grid(axis='y')  # Show gridlines
plt.axhline(0, color='black', linewidth=2, label='Y=0') 
plt.legend()

plt.savefig('line_chart_macd_and_signal.png')
```
